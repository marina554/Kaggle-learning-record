```python
# ===============================
# Titanic 初級モデル（Sex + Pclass + Age）
# FutureWarning 回避バージョン
# ===============================

import pandas as pd
from sklearn.ensemble import RandomForestClassifier

# データ読み込み
train = pd.read_csv('/kaggle/input/titanic/train.csv')
test = pd.read_csv('/kaggle/input/titanic/test.csv')

# ===============================
# データ前処理
# ===============================

# Ageの欠損値を中央値で補完（inplaceを使わず代入）
train['Age'] = train['Age'].fillna(train['Age'].median())
test['Age'] = test['Age'].fillna(train['Age'].median())  # trainの中央値で統一

# Sexを数値に変換（female=1, male=0）
train['Sex'] = train['Sex'].map({'female':1,'male':0})
test['Sex'] = test['Sex'].map({'female':1,'male':0})

# 特徴量と目的変数を設定
features = ['Sex','Pclass','Age']
X = train[features]
y = train['Survived']

# ===============================
# モデル作成・学習
# ===============================
model = RandomForestClassifier(random_state=42)
model.fit(X, y)

# ===============================
# テストデータで予測
# ===============================
test_X = test[features]
test['Survived'] = model.predict(test_X)

# ===============================
# 提出用ファイル作成
# ===============================
submission = test[['PassengerId','Survived']]
submission.to_csv('/kaggle/working/submission.csv', index=False)

print("提出ファイル submission.csv を作成しました！（FutureWarningなし）")
```
